```python
1: What is hypothesis testing in statistics?

ans- Hypothesis testing in statistics is a formal procedure used to evaluate competing claims or ideas about a population parameter.
It involves formulating a null hypothesis and an alternative hypothesis, collecting data, and using statistical methods to determine
whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.
```

2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?

ans-
The null hypothesis ($H_0$) is a statement of no effect, no difference, or no relationship. It represents the status quo or the claim being tested against.

The alternative hypothesis ($H_1$ or $H_a$) is a statement that contradicts the null hypothesis. It represents the claim that the researcher is trying to find evidence for. It can be one-sided (specifying a direction, e.g., greater than or less than) or two-sided (simply stating there is a difference).

The key difference lies in their purpose within hypothesis testing:

*   **Null Hypothesis ($H_0$):** Assumed to be true until evidence suggests otherwise. The goal is to see if the data provides enough evidence to *reject* this assumption.
*   **Alternative Hypothesis ($H_1$ or $H_a$):** The claim the researcher hopes to support. It's accepted only if there is sufficient evidence to reject the null hypothesis.

3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

ans- The significance level ( α ) is a pre-determined threshold used in hypothesis testing to decide whether to reject the null hypothesis. It represents the probability of rejecting the null hypothesis when it is actually true (Type I error).

Commonly used significance levels are 0.05 (or 5%), 0.01 (or 1%), and 0.10 (or 10%). A lower significance level means less willingness to reject the null hypothesis.

role in deciding the outcome:

Compare p-value to  α : After conducting a statistical test, a p-value is calculated. The p-value is the probability of observing the data (or more extreme data) if the null hypothesis is true.
Decision Rule:
If the p-value is less than or equal to  α  (p-value  ≤α ), we reject the null hypothesis. This suggests that the observed data is unlikely to have occurred by random chance if the null hypothesis were true, providing evidence in favor of the alternative hypothesis.
If the p-value is greater than  α  (p-value  >α ), we fail to reject the null hypothesis. This means that the observed data is reasonably likely to have occurred by random chance even if the null hypothesis were true, and there is not enough evidence to support the alternative hypothesis.


4: What are Type I and Type II errors? Give examples of each.

ans- In hypothesis testing, two main types of errors can occur:

Type I Error (False Positive): This occurs when you reject the null hypothesis ( H0 ) when it is actually true. The probability of making a Type I error is equal to the significance level ( α ).

Example: A medical test for a disease has a Type I error if it indicates a person has the disease when they actually don't. In hypothesis testing terms,  H0 : The person does not have the disease. Rejecting  H0  (saying they do have the disease) when it's true that they don't is a Type I error.
Type II Error (False Negative): This occurs when you fail to reject the null hypothesis ( H0 ) when it is actually false (i.e., the alternative hypothesis ( H1 ) is true). The probability of making a Type II error is denoted by  β .

Example: Using the same medical test, a Type II error occurs if the test indicates a person does not have the disease when they actually do. In hypothesis testing terms,  H0 : The person does not have the disease. Failing to reject  H0  (saying they don't have the disease) when  H0  is false (they do have the disease) is a Type II error.


5: What is the difference between a Z-test and a T-test? Explain when to use
each.

ans- Both Z-tests and T-tests are statistical tests used to compare means and make inferences about populations based on sample data. The main difference lies in the information we have about the population standard deviation ( σ ).

Z-test:

uses:-

Z-test:

It is used when you know the population standard deviation ( σ ). This is often the case when dealing with large sample sizes (generally  n≥30 ) or when the population standard deviation is known from previous studies or theoretical considerations.
Distribution: It uses the standard normal distribution (Z-distribution).
Formula: The test statistic is calculated as  Z=(x̄ −μ)/(σ/n−−√) , where  x̄   is the sample mean,  μ  is the population mean,  σ  is the population standard deviation, and  n  is the sample size.


T-test:

When to use: A T-test is used when you do not know the population standard deviation ( σ ) and have to estimate it using the sample standard deviation ( s ). This is common with smaller sample sizes (generally  n<30 ).
Distribution: It uses the t-distribution, which has heavier tails than the normal distribution to account for the uncertainty introduced by estimating the population standard deviation from a sample. The shape of the t-distribution depends on the degrees of freedom, which are related to the sample size ( n−1  for a one-sample t-test).
Formula: The test statistic is calculated as  T=(x̄ −μ)/(s/n−−√) , where  x̄   is the sample mean,  μ  is the hypothesized population mean,  s  is the sample standard deviation, and  n  is the sample size.

 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

mu = 100
sigma = 15
sample_size = 200
normal_data = np.random.normal(mu, sigma, sample_size)

confidence_level = 0.95
degrees_freedom = sample_size - 1

sample_mean = np.mean(normal_data)
sample_standard_error = stats.sem(normal_data)

confidence_interval = stats.t.interval(confidence_level, degrees_freedom,
                                       loc=sample_mean, scale=sample_standard_error)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")

# Plot the data
plt.figure(figsize=(10, 6))
plt.hist(normal_data, bins=20, density=True, alpha=0.6, color='g')
plt.title('Histogram of Simulated Normal Distribution Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_and_plot_zscores(data):
  """
  Calculates the Z-scores for a dataset and visualizes the standardized data.

  Args:
    data: A list or NumPy array of numerical data.
  """
  # Calculate the mean and standard deviation
  mean = np.mean(data)
  std_dev = np.std(data)

  # Calculate the Z-scores
  z_scores = (data - mean) / std_dev

  print(f"Original Data Mean: {mean:.2f}")
  print(f"Original Data Standard Deviation: {std_dev:.2f}")
  print(f"Z-scores Mean: {np.mean(z_scores):.2f}")
  print(f"Z-scores Standard Deviation: {np.std(z_scores):.2f}")

  # Plot the histogram of Z-scores
  plt.figure(figsize=(10, 6))
  sns.histplot(z_scores, kde=True, color='purple')
  plt.title('Histogram of Z-scores (Standardized Data)')
  plt.xlabel('Z-score')
  plt.ylabel('Frequency')
  plt.grid(axis='y', alpha=0.75)
  plt.show()

# Example usage with the sample data from the Z-test question
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

calculate_and_plot_zscores(sample_data)
```

 Z-scores:-


Z-scores, also known as standard scores, are a way to standardize data from different distributions so that they can be compared on a common scale.

What Z-scores represent:

A Z-score for a particular data point tells you how many standard deviations away from the mean that data point is.

A Z-score of 0 means the data point is exactly at the mean.
A positive Z-score means the data point is above the mean. For example, a Z-score of 1 means the data point is one standard deviation above the mean.
A negative Z-score means the data point is below the mean. For example, a Z-score of -2 means the data point is two standard deviations below the mean.
Why are Z-scores useful?

Comparison: They allow you to compare data points from datasets with different means and standard deviations.
Outlier detection: Data points with very large positive or negative Z-scores (typically beyond +3 or -3) are often considered outliers.
Standardization: They transform a dataset to have a mean of 0 and a standard deviation of 1. This is useful for many statistical techniques and machine learning algorithms.
The histogram of the Z-scores will show the distribution of your data in terms of standard deviations from the mean. You'll notice that the center of the histogram will be around 0, and the spread of the data will be based on standard deviations.
